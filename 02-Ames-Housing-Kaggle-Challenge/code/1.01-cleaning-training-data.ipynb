## Step 1: Dealing with Missing Data
We are going  to check for missing data values in each of our columns. Starting at the most problematic columns.

We will use MissingNo to visualize any missing data. Our goal is to have the below graphic totally black.

```python
# Check for missing values in each row
msno.matrix(df);
plt.title("Missing Data From Training Data Set",size=32);
```

```python
# Check for missing values in each row and sort by the most problematic.
def pull_bad_cols(df=df):
    '''
    This function simply runs through our dataframe and check columns for the percentage of missing data. 
    It then partitions them into sets of decreasing "troublesomeness".
    This creates a sort of "target list" for manageing our data.
    '''
    # bad_cols 
    bad_cols = df.isnull().mean().sort_values(ascending=False)
    return bad_cols[bad_cols > 0]

pull_bad_cols()
```

### Targeting Numeric Columns

We will start our data cleaning with imputing values into the numeric columns. Hopefully some of these will be zero, and we will use a regression to impute the rest.

```python
# num_cols are all the bad_cols with numeric data types
# We will be replacing the null values with numbers
bad_cols = pull_bad_cols().index

num_cols = df[bad_cols].select_dtypes(include='number').columns
df[num_cols].describe().T
```

#### Impute Zero to Num Cols with `.min()` = 0

```python
# Impute missing values with 0 for numeric columns whose minimum values are 0.
num_nulls = df[num_cols].min()
impute_zero_cols = (num_nulls[num_nulls == 0]).index.values
impute_reg_cols = (num_nulls[num_nulls != 0]).index.values
# print(num_cols)

print("We're going to put zeroes into these nulls:",impute_zero_cols)
print("We are going to get fancy with these ones:",impute_reg_cols)
```

```python
# Fill all NaN values with 0 and check the result.
df[impute_zero_cols] = df[impute_zero_cols].fillna(0.0)
df[impute_zero_cols].isnull().mean()
```

#### Imputing values in other columns.

```python
df[impute_reg_cols].isnull().mean()
```

#### Garage Yr Built = Yr Built
We will insert the year the house was built for values  missing the Garge Year Built date.

```python
# I want to make sure  my column update works correctly.
# So I am pulling the rows with a null year value
garage_yr_null = df['garage_yr_blt'].isnull()

# I am also pulling some control rows so that I am sure I don't overwrite good data.
garage_control = df['garage_yr_blt'].between(2000,2010)

# I will run this afterwards to make sure things  checked out alright.
df.loc[garage_yr_null | garage_control,['garage_yr_blt','year_built']]
```

```python
# Update columns of Garage Yr Built with values from Year built, where garage values are null.
df['garage_yr_blt'] = df['garage_yr_blt'].where(~garage_yr_null,df['year_built'])
```

```python
df.loc[garage_yr_null | garage_control,['garage_yr_blt','year_built']]
```

```python
df['garage_yr_blt'].isnull().mean()
```

#### Imputing Lot Frontage based on Lot Area

```python
df[impute_reg_cols].isnull().mean()
```

```python
df[df['lot_frontage'].isnull()]['ms_zoning'].value_counts()
```

Lot frontage is the length of road bordering the property. I wanted to make sure the properties without Lot Frontage were not apartment buildings or some sort of residence without street access.

The majority of these residences are RL: Residential Low Density, defined in the data dictionary as 'less than 5 residences per acre'.

Since these houses will defintiely have lot frontage, I am going to create a regression with the Lot Area. These two values are related, so the imputation will be, at least, reasonable.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
lot_null = ~df['lot_frontage'].isnull()
lr = LinearRegression()
X = np.log1p(df.loc[lot_null,['lot_frontage']])
y = np.log1p(df.loc[lot_null,'lot_frontage'])
```

```python
cross_val_score(lr, X, y, cv=5).mean()
```

> It's not a great model.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
lot_area_log = np.log1p(df[['lot_area']])
lot_front_pred_log = lr.predict(lot_area_log)
lot_front_exp = np.expm1(lot_front_pred_log)

lot_fill_values = pd.Series(lot_front_exp)
df['lot_frontage'] =  df['lot_frontage'].where(lot_null, lot_fill_values)
```

```python
# Check that our null values are gone
df['lot_frontage'].isnull().mean()
```

#### Numeric Columns Complete
We have filled all of our null values for each numeric column.

### Quality Scales
It looks like columns `'Pool QC'` and `'Fireplace Qu'` values are put on a  similar **qualitative scale**. The `'Pool QC'` has the qualitative values below:

|Code|Quality|
|---|---|
|Ex|Excellent|
|Gd|Good|
|TA|Average/Typical|
|Fa|Fair|
|NA|No Pool|

Entries for 'NA' are being interpreted as `np.nan` values, which is resulting in our high percentage of nulls. This scale can be converted into a  numeric scale using the information in the [variable descriptions](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt) page. 
We can fill the null values and convert the `'Pool QC'` and `'Fireplace Qu'` columns to numeric scales.

We will also look for **any other columns** in the data frame that can be corrected  in this way. We  will save these columns  in a list called `qual_cols` and convert them using the `.map()` method and the the `quality_dict` dictionary below.

```python
qual_dict = {
        'Ex':5,
        'Gd':4,
        'TA':3,
        'Fa':2,
        'Po':1,
        'None':0
    }

# This is a regular expression that will help me find any and all columns that might be convertible into a quality scale.
qual_str = "(^" + "$|^".join(qual_dict.keys()) + "$)"
qual_str
```

```python
qual_cols = []
for col in df.columns:
    if df[col].dtype == 'object':
        if df[col].str.contains(r"(^Ex$|^Gd$|^TA$|^Fa$|^Po$)",regex=True).sum():
            qual_cols.append(col)
qual_cols
```

#### Filling `nan` values with 'None' in  'Quality' Columns

This is similar to many other qualitative columns in our data. We are going to create a `quality_dict`, to hold our quality-quantitative pairs and the function `qual_to_quant` to apply the conversions.

```python
# Fill the  column's null values with 'None' so that we can apply the dictionary.
for col in df[qual_cols]:
    df[col].fillna('None',inplace=True)
    print(col,df[col].unique())
```

Since `'bsmt_exposure'` does not have the same qualitative scale, we will deal with it later using a separate dictionary.

```python
qual_cols.remove('bsmt_exposure')
```

#### Converting the qualitative values to quantitative using `.map(qual_dict)`
Here we convert the  qualitative values of our columns to a  numeric scale usign the `.map()` and `qual_dict` dictionary.

```python
for col in qual_cols:
    df[col] = df[col].map(qual_dict)
df[qual_cols]
```

#### DONE ! Qualitative Columns:  Nulls Filled and Numerically Converted
We have cleaned 10 columns of data plus all the numeric columns.

```python
pull_bad_cols()
```
