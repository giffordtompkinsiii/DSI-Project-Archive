# Preprocessing and Feature Engineering

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/model.csv')
```

```python
df['genre'] = df['genre'].fillna('None')
```

```python
genres = set(' '.join([str(genre) for genre in df['genre'].values]).split(' '))
```

```python
df['genre_list'] = df['genre'].str.split(' ')
```

```python
df['genre'].sort_values()
```

```python
for genre in genres:
    print(genre.title())
    df[genre] = df['genre_list'].map(lambda x: 1 if genre in x else 0)
df
```

```python
df['status'] = df['status'].map(lambda x: 1 if x == 'Renewed' else 0)
```

```python
df['avg_length'] = (df['length_hi'] + df['length_low'])/2
```

```python
df.rename({'status':'renewed'}, axis=1, inplace=True)
```

```python
df.corr()['status'].sort_values(ascending=False)
```

```python
df = pd.get_dummies(df, columns=['premiere_month'])
```

```python
df.to_csv('./data/processed_data.csv', index=False)
```
